```python
import matplotlib.pyplot as plt
import pandas as pd
from collections import defaultdict
```

# Read in data

```python
infile = "../Assignment/importexport.txt"

exD = defaultdict(list)
imD = defaultdict(list)

with open(infile, "r") as fin:
    next(fin)
    for line in fin:
        cols = line.rstrip("\n").split("\t")
        exD[cols[0]].append(float(cols[4]))
        imD[cols[0]].append(float(cols[3]))

print(len(exD), len(imD))
```

# Generate Plot

```python
fig,ax = plt.subplots(1,1,figsize=(4,4))
ax.scatter(imD["WesternUrban"], exD["WesternUrban"], label="Western Urban", c="#e34a33", alpha=0.5)
ax.scatter(imD["WesternRural"], exD["WesternRural"], label="Western Rural", c="b", alpha=0.5)
ax.set_xlabel("Number of Imports", fontsize=20)
ax.set_ylabel("Number of Exports", fontsize=20)
ax.legend()

fig.savefig("Demo1_scatter.png", dpi=300, bbox_inches="tight")
```

# Redo figure, but using pandas

```python
df = pd.read_csv(infile, sep="\t", header=0)
wu = df[df["District"]=="WesternUrban"]
wr = df[df["District"]=="WesternRural"]
```

```python
fig,ax = plt.subplots(1,1,figsize=(4,4))
ax.scatter(wu["Imports"], wu["Exports"], label="Western Urban", c="#e34a33", alpha=0.5)
ax.scatter(wr["Imports"], wr["Exports"], label="Western Rural", c="b", alpha=0.5)
ax.set_xlabel("Number of Imports", fontsize=20)
ax.set_ylabel("Number of Exports", fontsize=20)
ax.legend()

fig.savefig("Demo1_scatter_Pandas.png", dpi=300, bbox_inches="tight")
```

# Generate figure with two panels

```python
fig,axes = plt.subplots(1,2,figsize=(9,4), sharey=True, sharex=True)
axes[0].scatter(wu["Imports"], wu["Exports"], label="Western Urban", c="#e34a33", alpha=0.5)
axes[1].scatter(wr["Imports"], wr["Exports"], label="Western Rural", c="b", alpha=0.5)

for i in [0,1]:
    axes[i].set_xlabel("Number of Imports", fontsize=20)
axes[0].set_ylabel("Number of Exports", fontsize=20)
#axes.legend()

fig.savefig("Demo1_scatter_MultiPanel.png", dpi=300, bbox_inches="tight")
```

# Generate a histogram

```python
fig,ax = plt.subplots(1,1,figsize=(4,4))
ax.hist(wu["Imports"], label="Western Urban", color="#e34a33", alpha=0.8, bins=10, zorder=2)
ax.hist(wr["Imports"], label="Western Rural", color="b", alpha=0.8, bins=20, zorder=1)
ax.set_xlabel("Number of Imports", fontsize=20)
ax.set_ylabel("Frequency", fontsize=20)
ax.legend()

fig.savefig("Demo1_hist.png", dpi=300, bbox_inches="tight")
```
