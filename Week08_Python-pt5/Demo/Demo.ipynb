```python
# Make a new dataframe that only contains the negative control samples

negDF = pd.DataFrame()
for col in df2.columns:
    if col.startswith("Neg"):
        negDF[col] = df2[col]
```

```python
negDF.head()
```

```python
#Get total read counts for each assay

sums = df2.sum(axis=0)
sums
```

```python
# Normalize the individual counts = relative abundance

norm = df2/sums*1000000
norm.head()
```

```python
# Generate average across replicates
    # Returns a Series

avg = (norm["BB001_A"] + norm["BB001_B"])/2
avg.head()
```
